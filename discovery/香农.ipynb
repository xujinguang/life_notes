### 香农极限
01信息输出物理媒介上会有噪音，比如静电，电磁场
1. 乱序
2. 置换
3. 缺失
4. 噪声
因此在物理信道上，有没有一个速度限制？这就是理论上传输速度极限，好比光速最大原理，所有事物的速度都不会超过光速一样。当然不同的是光速无法突破，但是信息传输可以超过，但是结果是无法还原原信息成为一堆垃圾噪声
